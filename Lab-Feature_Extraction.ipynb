```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Variable objetivo
target_variable = 'Customer Lifetime Value'

# Variables categóricas
categorical_df = customer_df.select_dtypes(include=['object', 'category'])

# Configuración de los gráficos
plt.figure(figsize=(15, 10))

# Crear un gráfico de caja para cada variable categórica
for i, column in enumerate(categorical_df.columns, 1):
    plt.subplot((len(categorical_df.columns) + 2) // 3, 3, i)
    sns.boxplot(x=customer_df[column], y=customer_df[target_variable])
    plt.title(f'{column} vs {target_variable}')
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Variables categóricas
categorical_df = customer_df.select_dtypes(include=['object', 'category'])

# Verificar la representación de las diferentes categorías en cada columna
for column in categorical_df.columns:
    print(f'Columna: {column}')
    print(customer_df[column].value_counts(dropna=False))
    print('\n')
```

Observaciones Potenciales en los Gráficos:
Distribución del Objetivo: Ver cómo varía la distribución del Customer Lifetime Value entre las diferentes categorías de cada variable categórica.
Outliers: Identificar posibles valores atípicos (outliers) en los datos.
Comparaciones entre Categorías: Comparar fácilmente las medianas, cuartiles y rangos de los valores del objetivo entre diferentes categorías.
Categorías con Baja Representación: Identificar categorías con muy pocas observaciones que podrían necesitar ser agrupadas o tratadas de otra manera.
